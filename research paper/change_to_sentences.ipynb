Better Sentence-Level Segmentation in Python
pySBD: Python Sentence Boundary Disambiguation (SBD):

```python
import pysbd
import pandas as pd
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer
import ast
import numpy as np

```

```python
bt_speeches = pd.read_csv("C:/Users/rober/UCD/pol42050/new-sound-of-politics/data/bundestag_cleaned_period20.csv")
seg = pysbd.Segmenter(language="de", clean=False)
```

```python
bt_speeches["Speech"] = [seg.segment(speech) for speech in bt_speeches["Speech"]]
```

```python

# Load data
# bt_speeches = pd.read_csv("C:/Users/rober/UCD/pol42050/new-sound-of-politics/data/bundestag_sentences20.csv", index_col="Unnamed: 0")
# len(bt_speeches["Speech"])
# bt_speeches20 = bt_speeches[bt_speeches["Period"] == 20]
# bt_speeches20.to_csv("C:/Users/rober/UCD/pol42050/new-sound-of-politics/data/bundestag_sentences20.csv")
```

```python
# bt_speeches = bt_speeches.sample(frac=0.001)

bt_speeches['Speech'] = bt_speeches['Speech'].apply(ast.literal_eval)

```

```python
# Explode Speech lists into sentences with speech_id and position
exploded_data = []
for idx, row in bt_speeches.iterrows():
    sentences = row['Speech']
    for pos, sentence in enumerate(sentences, 1):
        exploded_data.append({
            'speech_id': idx,
            'sentence_position': pos,
            'sentence': sentence
        })

# Create new DataFrame with exploded sentences
bt_sentences = pd.DataFrame(exploded_data)

# Map original columns to new structure
bt_sentences['MPID'] = bt_sentences['speech_id'].map(bt_speeches['MPID'])
bt_sentences['Date'] = bt_sentences['speech_id'].map(bt_speeches['Date'])
bt_sentences['Chair'] = bt_sentences['speech_id'].map(bt_speeches['Chair'])
bt_sentences['Session'] = bt_sentences['speech_id'].map(bt_speeches['Session'])
bt_sentences['Period'] = bt_sentences['speech_id'].map(bt_speeches['Period'])
bt_sentences['Party'] = bt_sentences['speech_id'].map(bt_speeches['Party'])
bt_sentences['Name'] = bt_sentences['speech_id'].map(bt_speeches['Name'])

# Reorder columns for clarity (optional)
column_order = ['speech_id', 'sentence_position', 'sentence', 'MPID', 'Date', 'Chair', 
                'Session', 'Period', 'Party', 'Name']
bt_sentences = bt_sentences[column_order]

# Drop original Speech column if desired (optional)
# bt_sentences = bt_sentences.drop(columns=['Speech'])
```

```python

bt_sentences.to_csv("bundestag_sentences20.csv", index=False)
```
